```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#importing data sets
instacart_orders = pd.read_csv('../instacart_2017_05_01/orders.csv')
instacart_products = pd.read_csv('../instacart_2017_05_01/products.csv')
instacart_dept = pd.read_csv('../instacart_2017_05_01/departments.csv')
instacart_order_products__train = pd.read_csv('../instacart_2017_05_01/order_products__train.csv')
```

```python
#create and merge data frames so I can analyze across Product IDs x Order IDs
products_df1 = pd.DataFrame(instacart_products)
orders_df2 = pd.DataFrame(instacart_orders)
dept_df3 = pd.DataFrame(instacart_dept)
orders_products_train_df3 = pd.DataFrame(instacart_order_products__train)
df4 = pd.merge(products_df1, orders_products_train_df3)
df5 = pd.merge(df4, orders_df2)
df6 = pd.merge(df5, dept_df3)
df6.columns
```

```python
df6.loc[:, 'days_since_prior_order'].fillna(df6.loc[:, 'days_since_prior_order'].mean(), inplace=True)
```

```python
order_days_dummies = pd.get_dummies(df6.loc[:, 'days_since_prior_order'], prefix='days_since_prior_order')
order_days_dummies.drop(order_days_dummies.columns[0], axis=1, inplace=True)
```

```python
df6 = pd.concat([df6, order_days_dummies], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X, y)
```
