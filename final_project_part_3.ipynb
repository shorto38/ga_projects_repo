```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#importing data sets
instacart_orders = pd.read_csv('../instacart_2017_05_01/orders.csv')
instacart_products = pd.read_csv('../instacart_2017_05_01/products.csv')
instacart_dept = pd.read_csv('../instacart_2017_05_01/departments.csv')
instacart_order_products__train = pd.read_csv('../instacart_2017_05_01/order_products__train.csv')
```

```python
#merge datasets into dataframe to see all data
products_df1 = pd.DataFrame(instacart_products)
orders_df2 = pd.DataFrame(instacart_orders)
dept_df3 = pd.DataFrame(instacart_dept)
orders_products_train_df3 = pd.DataFrame(instacart_order_products__train)
df4 = pd.merge(products_df1, orders_products_train_df3)
df5 = pd.merge(df4, orders_df2)
df6 = pd.merge(df5, dept_df3)
df6.columns
```

```python
#how many orders per unique user-id
amt_orders_per_user = orders_df2.loc[:, 'user_id'].value_counts().sort_index()
amt_orders_per_user
```

```python
# generating amt_orders_per_user DF to plot how many orders per unique user
amt_orders_per_user = pd.DataFrame(df6.user_id.value_counts().reset_index())
amt_orders_per_user.columns = ['unique_user', 'order_count']
print(amt_orders_per_user)
```

```python
amt_orders_per_user.loc[:, 'order_count'].head(100).sort_index().plot(kind='bar', figsize=(15,15));
plt.xlabel('unique_user');
plt.ylabel('order_count');
plt.title('Orders per User');
```

```python
#the average amount of orders per user is 10.5
amt_orders_per_user.loc[:, 'order_count'].mean()
```
